```python
import glob
import pandas as pd
import utils
from sklearn import preprocessing
import numpy as np
```

```python
# Get all the files in the specified directory
path_to_csv_files = r'E:\CIC'
csv_file_names = glob.glob(path_to_csv_files + "\*.csv")

# Print friendly message to so user knows how many files are being read
num_of_files = len(csv_file_names)
print(f"Found {num_of_files} files")
```

```python
dfs = []
# Construct the pandas data frame into a single data frame
# Iterate through all the files and append the data frame to the array
for file_name in csv_file_names:
    df = utils.read_csv_chunked(file_name)
    dfs.append(df)
```

```python
# Concatenate all data into one data frame
data_df = pd.concat(dfs)

# Print the data frame columns
print('Number of columns: ', data_df.columns.size)
print('Number of rows and columns: ', data_df.shape)
```

```python
sorted(data_df.columns)
```

```python
# Remove the following columns from the dataset
# Remove  'Dst IP', 'Flow ID', 'Src IP', 'Src Port', since it's only available in Thursday-20-02
# Remove 'Timestamp' since it's not useful to calculating if a packet is malicious
remove_columns = ['Timestamp', 'Dst IP', 'Flow ID', 'Src IP', 'Src Port']
print('Removing headers: ', remove_columns)
data_df = data_df.drop(remove_columns, axis=1)

# Print the data frame columns
print('Number of rows and columns: ', data_df.shape)
print(data_df.columns)
```

```python
# Display the label column values
print('Before label encoder')
print('---------------------')
print(data_df['Label'].value_counts())
```

```python
# We want to replace all the attack labels with a generic "Malcious" label for our classifier problem
# That way it will result to "0" or "1" for benign or malicous respectfully
replace_labels = ["Bot", "FTP-BruteForce", "SSH-Bruteforce", "DoS attacks-GoldenEye", "DoS attacks-Slowloris"]

# This might be the complete list
# label_names = ['Benign','Bot','Brute Force-Web','Brute Force-XSS','DDOS attack-HOIC', 'DDOS attack-LOIC-UDP', 'DDoS attacks-LOIC-HTTP',
# 'DoS attacks-GoldenEye', 'DoS attacks-Hulk', 'DoS attacks-SlowHTTPTest', 'DoS attacks-Slowloris', 'FTP-BruteForce',
# 'Infilteration', 'Label', 'SQL Injection', 'SSH-Bruteforce']

data_df = data_df.replace(to_replace=replace_labels, value="Malicious")

# Print result of encoding (Should match above)
data_df['Label'].value_counts()
```

```python
# Instanstiate label encoder for use
label_encoder = preprocessing.LabelEncoder()

# Encode all strings to numerical values
data_df['Label'] = label_encoder.fit_transform(data_df['Label'].values)

# Print result of encoding (Should match above)
data_df['Label'].value_counts()
```
